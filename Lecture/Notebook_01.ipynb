<h1> 1. Jupyter Notebook Basics</h1>

<p>The Jupyter Notebook consists of 2 things:</p>

<ul>
<li> A kernel (in our case a Python interpreter but Jupyter can handle <a href="https://github.com/jupyter/jupyter/wiki/Jupyter-kernels">many other languages</a>).</li>
<li>A web interface.</li>

<p>The notebook consists of <b>cells</b> where you can enter text or perform computations. Let us have a look at them.</p>

<hr />
<h2>There are 3 main types of cells:</h2>

<ol>
<li><font color="FireBrick">Markdown cells</font>: Cells like this one where you can type
    text, $\LaTeX$ (a typesetting language used in science), or Markdown (it is a <a href="http://en.wikipedia.org/wiki/Markup_language">markup language</a>, a subset of HTML, the markup language that makes up the content of webpages).      
    
<b>Note:</b> A Markdown cell can also contain HTML code.</li><br>

<li><font color="FireBrick">Code</font> cells: These are cells where you enter Python code to be executed by the interpreter.</li><br>   
    
<li><font color="FireBrick">Raw cells</font>: Where the text is reproduced as is.<br>
These are useful for including a piece of code without having the notebook try to execute it.</li>
</ol>

<h1>2. Module imports</h1>

<ul>
<li> <font color="Red">math</font>: Basic mathematical functions and constants.</li>
<li> <font color="Red">os</font>: Interfaces with the operating system.</li>
<li> <font color="Red">cmath</font>: Functions of complex variables.</li>
<li> <font color="Red">numpy</font> [Numerical Python]: Similar to Matlab's matrix computation.</li>
<li> <font color="Red">scipy</font> [Scientific Python]: Scientific computing (e.g., ODE solution, numerical</li> quadrature).
<li> <font color="Red">pandas</font> [Data analysis]: We'll look at this in detail later.</li>
<li> <font color="Red">matplotlib</font>: Plotting library (similar to Matlab plotting).</li>
<li> <font color="Red">pathlib</font>: Handling of paths across operating systems.</li>    
<li> <font color="Red">IPython.display</font>: For fancy notebook content.</li>
</ul>

<p>and many others... See the <a href="http://docs.python.org/3/library/"><b>Python Standard Library</b></a>.</p>

<h3> <center><font color=MediumVioletRed>A. You can access the content of a module with the <code>import</code> command.</font></center></h3>
<h4> A module only needs to be imported <b>once</b> per notebook. </h4>
(unless you modify the content of the module)

As an example we will use the <b>math</b> module. 

By default, this module is not loaded so functions like the square root (<code>sqrt</code>) are not defined.

````python
# This line is a comment within a Python code cell.

sqrt(2)   # this line will raise an exception (error)
````

````python
# By importing the math module we have access to 
# all the python objects of the module, 
# including the square root function.

import math
````

#### All the variables and functions of the `math` module are now available to us.

````python
# To list the content of the math module

print(dir(math))

# Counting the number of names using the function len 
print(len(dir(math)))
````

<h3> <center><font color=MediumVioletRed>B. You can use <kbd>TAB</kbd> to get options for autocompletion and <kbd>?</kbd> to get help.</font></center></h3>


#### If you type `math` followed by a dot, and then press the `TAB` key on your keyboard, the interpreter will present you with possible completions. 

This is a very useful feature when you can't remember the names of all the functions or variables in a module. 

Try it below.

````python
math.
````

````python
# To get some information about a function, say log1p. 

help(math.log1p)
````

````python
# Another way to get help is to use a question mark.

math.log1p?
````

````python
# Let's calculate the square root again.
# The syntax with the "dot" means we are 
# calling the sqrt function of the math module.

math.sqrt(2)
````

<h3> <center><font color=MediumVioletRed>C. You can load as little or as much of a module as you need.</font></center></h3>


#### We can load all the content of the module using 

```python
    from math import *
```

The star means import everything (functions, variables) from the module (we are actually making the <em>namespace</em> of the math module available to our code).


````python
# We no longer have to use the math prefix.

from math import *
sqrt(2)
````

<br>
<div style="background-color:#FBEFFB;"><p style="font-size:20px;color:#FF0080">&#9888; Beware!</p> <!--- Warning --->

<p> This is not a good practice as it can cause overwriting function names. This occurs when two modules have functions with the same name. See an example of this below when importing the `sqrt` function from another module. </p> 


#### Note that the square root function of the `math` module doesn't handle complex numbers (written as `a + jb`)

````python
sqrt(1+2j)
````

#### To deal with complex arguments we need the `sqrt` function of the `cmath` module.

````python
import cmath

cmath.sqrt(1 + 2j)
````

````python
from cmath import sqrt

sqrt(1 + 2j)
````

<div style="background-color:#F7F2E0;">
<h3> <font color=MediumVioletRed>Remark:</font> </h3>
<ul>
  <li> We have now "clobbered" (overwritten) the standard square root function with the "complex" one.</li><br>
  <li> If all we need is a couple of functions, we do not have to import the entire module. We can simply import the function(s) we want. </li><br>
  <li> To avoid confusion with the "normal" <code>sqrt</code> function we also rename <code>cmath.sqrt</code> into <code>csqrt</code></li>
  </ul>  
</div>

````python
# Notice how by importing just the function 
# we don't need to write cmath.csqrt.
# In effect we've renamed the function from 
# cmath.sqrt to simply csqrt.
from math import sqrt
from cmath import sqrt as csqrt
sqrt(1), csqrt(-1)
````

#### Note that some modules are so large that it is not always practical or recommended to use "`import *`".

````python
import numpy
print(dir(numpy))
print(len(dir(numpy)))
````

#### It is better to either reference the full name of the function or import it explicitly.

For instance, if we want to use the `matrix` object of the `numpy` module, we could use:


```python 
import numpy
``` 

and then we can call `numpy.matrix()` (the full name).

```python
import numpy as np
```

and then we can call `np.matrix()` (the most common and preferred way).

```python
from numpy import matrix
```

and then call `matrix()` (this way is not advisable).
