Let us try to automate the loading of all the files.

First we need a list of all the files in our directory. 

A <code>Path</code> object has a <code>glob</code> method that returns a generator with the elements within a given directory.

```python
dataset_dir = Path('names/')
#print(list(dataset_dir.glob("*.txt")))
files = [item.name for item in sorted(dataset_dir.glob("*.txt"))]
files[:10]
```

```python
# Extract a list of years
years = [item[3:7] for item in files]
years[:10]
```

#### We gather the data into a single dataset in two steps:

1. We create a list of dataframes where each element corresponds to a particular year.
1. We concatenate all the data into a single dataframe.


```python
L = []
for file, year in zip(files, years):
    path = Path(dataset_dir, file)
    df = pd.read_csv(path, names=['name', 'sex', 'births'])
    df['year'] = year
    L.append(df)
```

<h3> <font color=MediumVioletRed>Note:</font> </h3>
<p>Note that <b>this is not</b> the number of babies of each sex with that name. </p>

<p>We've grouped the data by <code>year</code> and <code>sex</code> so this count corresponds to the <b>number of years</b> in which at least five babies of either sex were named Lucille (5 being the yearly threshold beyond which a name isn't recorded in the data set).</p>
</div>

If we wanted to compute the number of babies of each sex with that name we will have to sum the <code>births</code> variable. The first code below does this by year and sex. The second one by sex only.

```python
def sum_lucille_births(x):
    return x[x['name'] == 'Lucille']['births'].sum() 
    
names_groupby_sex_by_year = names.groupby(['sex', 'year'])

# by sex and year
display(names_groupby_sex_by_year.apply(sum_lucille_births).unstack(level = 0))

# by sex
display(names.groupby(['sex']).apply(sum_lucille_births))

# We could subset directly before grouping also do
display(names[names.name == 'Lucille'].groupby(['year', 'sex'])['births'].sum().unstack().fillna(0))
```

#### Let us put that code into a function to make it more convenient.

```python
def timeline_name(name='Basil', use_log=False, figsize=(20, 6)):
    ax = (names[names.name == name]
            .pivot_table('births', 
                         index='year', 
                         columns='sex', 
                         fill_value=0, 
                         aggfunc=np.sum)
            .plot(figsize=figsize,
                  logy=use_log, 
                  marker='o', 
                  lw=1,
                  color=('r', 'g'), 
                  alpha=0.5)
    )
    ax.set_xlabel('year', fontsize=14)
    ax.set_ylabel('count', fontsize=14)
    ax.set_title(name, fontsize=16)
```

```python
timeline_name('Lucille', use_log=True)
```

```python
timeline_name('Sylvester', use_log=True)
```

```python
timeline_name('Hannah', use_log=True)
```

<h3> <center><font color=MediumVioletRed>C. The Seinfeld effect?</font></center></h3>

<p>Note that a plausible explanation of why the dominant gender for a given name may have changed over the years can sometimes be found relatively easily, as in the case of the name <b>Ashley</b>... For other names, a change in popularity may be due to a less conventional reason... </p>

<p>In an episode of <a href="http://www.imdb.com/title/tt0697774/">Seinfeld</a> George mentions to his friends that should he ever have a kid, he would name him (or her) Seven, after the number of Yankee baseball player Mickey Mantle.</p>

<p>Let's see if the data can tell us anything about this.</p>

```python
'Seven' in names.name.unique()
```

The Seinfeld episode was aired in 1996.  

Let us see if there is any indication that it may have influenced baby names.

First let us create a date range over which to plot the data.

```python
seinfeld_year = '1996'
start = '1991'
end   = '2019'

# Let's create a range of dates from start to end.
date_range = pd.date_range(start, end, freq='A-JAN')
```

```python
date_range
```

```python
# The data
data = (names[names.name == 'Seven']
            .pivot_table('births', 
                         index='year', 
                         columns='sex', 
                         fill_value=0, 
                         aggfunc=np.sum)
       )

# The base plot
ax = data.plot(figsize=(16, 6), 
          logy=False, 
          marker='o', 
          color=('r', 'g'), 
          alpha=0.5, 
          title='Seven', 
          grid=False)

# The vertical strip
ax.axvline(seinfeld_year, ymin=0, ymax=1000, lw=15, color='orange', alpha=0.5)
ax.set_xticks(date_range)
ax.set_xlim([start, end])

# Ensure that the labels on the x axis match the years.
ax.set_xticklabels(date_range.year, rotation=0, ha='center' )

# Annotate the figure with some text by specifying location using xy parameter
ax.annotate('Episode "The Seven" is aired', 
            xy=(pd.to_datetime(seinfeld_year, format="%Y"), 40),
            xycoords='data',
            rotation=90,
            horizontalalignment='center'
            )

plt.show()
```

<p>Note that this does not prove that the show "created" the name fad (correlation &#8800; causation and all that), but it does suggest that it may have enhanced it...</p>

<p>We have looked at a few name trends over the years, and for some we can observe sharp increases, or decreases in the number of babies.</p> <p>However, this can be misleading. The number of babies with a given name in a given year is one thing, but in order to get an accurate view of the popularity of the name we should account for the fact that the total number of births changes over time.</p> 
<p>Therefore, we should normalise the number of boys and girls with a given name in a given year by the total number of births in that year. This way we can plot the proportion of boys or girls with a given name in that year.</p>

#### Let us get a list of the top 1000 most popular names in each year and for each sex.

#### We then use it to visualise the popularity of our chosen names across the years.

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'sex']) 
top1000 = grouped.apply(get_top1000).drop(['year', 'sex'], axis='columns')
top1000.head()
```

```python
# get rid of the original index
top1000.index = top1000.index.droplevel(2)
top1000.head()
```

```python
top1000.count()
```

```python
boys, girls = top1000[top1000.index.get_level_values(1) == 'M'],  top1000[top1000.index.get_level_values(1) == 'F']
```

```python
boys.head()
```

```python
girls.head()
```
