<h1> 2. Python functions</h1>

In this section we learn how to create our own functions.

Python implements a number of mathematical functions such as <code>sine</code> of <code>cosine</code>
but we are talking about more general <b>rules</b> that produce a specific output when given a suitable input.
Functions allow you to re-use complex operations without having to write the whole thing again.

Let us start with something simple. We want to create a function <code>sine_of_cosine</code> that will compute
$\sin[\cos(x)]$.

We'll write the function and we will comment on the syntax afterwards.


````python
from math import * # so we can use the trig functions and maths constants like pi.

def sine_of_cosine(x):
    return sin(cos(x))
````

<p>Processing the previous cell does not output anything. We have not <b>called</b> the function yet. We have just defined it.</p>

<h4> <font color=MediumVioletRed> Remark:</font> A function definition always starts with</h4>

```python
    def name_of_the_function(var1, var2, ...):
        
        body of the function...
```        
After the name of the function we have the input variable(s) between brackets.
If the function does not have any input variable, we would leave the brackets empty.

After the closing bracket there is a **colon** ":". 

Everything below the first line is the **body** of the funtion. Unlike other languages like C, Python does not use
curly brackets to determine where the body of the function starts and end. Instead
it uses **indentation**.

The convention for function names (and variables) in Python is to use lowercase letters with words separated by an underscore

<br>
<div style="background-color:#FBEFFB;"><p style="font-size:20px;color:#FF0080">&#9888; Beware!</p> <!--- Warning --->

<p>Python uses indentation to determine <b>scope</b>, that is what part of the code a particular statement belongs to (can be "seen" from). It does not matter if you use the TAB key or the SPACE bar to indent your code as long as you are consistent. We will be using the TAB key.</p>
</div>

````python
def say_hello():
    # This is a comment inside the function
    print("Hello from inside the function")
    return

# We are no longer inside the function
print("Hello from outside the function")
````

#### It is worth noting that the `print` function inside the function output did not print out the message, because we have not **called** the function.

````python
say_hello()  # call to the function which executes the code it contains.
print("Hello from outside the function")
say_hello()
````

#### Since we have already _declared_ our function `sine_of_cosine` we can now use it.

````python
print(sine_of_cosine(0))
````

````python
# You can use function composition and assign the result to a variable (provided the function returns something)
result = sine_of_cosine(sine_of_cosine(1))
print(result)
````

#### Functions are not limited to numerical parameters or outputs.

Let's create a function that takes 2 strings and repeats them $n$ times in an alternating pattern.

````python
def string_repeat(s1, s2, n = 2):   # You can provide a default value for your input variables
    s = n * (s1 + s2)
    return s

a = "Hello "
b = "World, "

# If we don't specify the value of n, the default value is used.
result = string_repeat(a, b)  # n=2 is used inside the function
print(result)
````

````python
result = string_repeat(a, b, 5)
print(result)
````

````python
# Any valid string can be passed to the function
result = string_repeat(b, b + a, n = 5)
print(result)

result = string_repeat(a, "Another string")
print(result)
````

<h4> More about <font color="Red"> scope</font>.</h4>

Recall the function we've just defined:

```python
    def string_repeat(s1, s2, n = 2):   
        s = n * (s1 + s2)
        return s
```
The variable <b>s</b> is called a <b>local</b> variable, it is only defined within the 
body of the function.

It is only defined as long as the function is running. 
It isn't defined *outside* of the function.

````python
def string_repeat(s1, s2, n = 2):   
    s = n * (s1 + s2)
    return s

print(s)   # will produce an error
````

````python
# Local variables do not clash with global ones

a = 10
def my_function():   
    a = 45
    return 

# Call the function
my_function()

print(a)   
````

<h3> <font color=MediumVioletRed>Remark:</font> </h3>

<p>Everything in Python is an object. This of course includes functions. This means that you can pass a function as an argument to
another function.</p> <p>We say that functions are first citizens in the language, or that Python has higher-order functions</p>


<p>Here is a simple example, a function that given an integer $n$ returns the function $f(x)=x^n$. </p>

<p>In other words we have a function that takes an integer and <b>returns a function</b>.
$$g(n): n\longrightarrow \{f(x)= x^n\}$$</p>

<p>In the jargon of computer science this is called a <b>closure</b>, but all it is, is 
a function that wraps and returns (outputs) another function who keeps a reference to the variables that are
defined in the outer function.</p>
</div>

````python
def my_power(n):
    def f(x):
        return x**n
    return f

# same as f(x) = x^2
square = my_power(2)
# same as f(x) = x^3
cube = my_power(3)

print(square(2), cube(2))
````

The function returned by <code>my_power</code> keeps a reference to the value of <code>n</code> that was passed to <code>my_power</code> when it was called.

### Function with a variable number of _unnamed_ arguments.

In python you'll often see the following pattern:

```python
def my_function(*args):
    # body of the function
```

The `*args` uses the star operator to unpack a tuple of arbitrary length. This is useful when you need to pass a variable number of arguments to the function.

````python
def my_function(*args):
    print(len(args))
    print(args)
````

````python
my_function(1, 2, 3)
````

````python
my_function(1, 2, 'A', 78, ['a', 'b'])
````

### Function with a variable number of _named_ arguments.
Similarly, you will often see the following pattern:

```python
def my_function(**kwargs):
    # body of the function
```

The `**kwargs` uses the double star operator to unpack a dictionary where the keys are the function parameters' names and the values are, well, their values.

````python
def my_function(**kwargs):
    print(kwargs)
````

````python
my_function(a=1, n=12, debug=True)
````

### Of course, the two can be combined to provide the most general interface to a function.

Keep in mind that Python expects all named parameters to come after all unnamed parameters.

````python
def my_function(*args, **kwargs):
    print('Unnamed parameters:')
    print(*args, sep='\n')
    print('Named parameters:')
    print(kwargs)
````

````python
my_function(2, 3, False, a=50, answer="Yes")
````

<h3> <center><font color=MediumVioletRed>D. <code>try</code> / <code>except</code> statement.</font></center></h3>

You may have noticed that when you make a mistake in your code, Python is not shy about letting you know about
it.
For instance, earlier when we were trying to output the value of a local variable outside of the function where it is defined, we got the following message:


```python
    ---------------------------------------------------------------------------
    NameError                                 Traceback (most recent call last)
    <ipython-input-41-3b703d77dd0d> in <module>()
          3     return s
          4 
    ----> 5 print(s)   # will produce an error

    NameError: name 's' is not defined

```


The Python interpreter has thrown an __exception__ to signal that an error occured. It is even kind enough to let
you know the type of error, here a `NameError`.


Python has many builtin types of errors and you can find them [here](https://docs.python.org/3.7/library/exceptions.html)
(Note that you can also define your own types).

For instance if you tried to divide by 0 you'd get a `ZeroDivisionError`.

````python
def my_function(a):
    return 2/a

my_function(0)
````

For a simple example like the one above, the error is not such a big deal, you can go back and fix it. However, imagine if you wrote
a piece of software where _someone other than you_ will be providing the value of variable `a`, or 
if you had an application with thousands of lines of codes where one function, deep within the code, requires that only non-zero
variables be passed to it. 


How do you prevent your code from crashing if a `ZeroDivisionError` occurs?

You may be tempted to use an `if else` construct and, for simple cases it may work well enough. For instance we could write.


````python
def my_function2(a):
    if (a != 0):
        return 2/a
    else:  
        return "You must pass a non-zero argument"
    
my_function2(0)
````


This looks fine. Problem solved, right? Not quite. 

Remember that you may know what type of variable must be passed to the function `my_function2` but Python does not know in advance since there is no compilation (Note: type hints would actually help here). 

This means that there is nothing stopping someone
using your code from doing something like:


````python
my_function2("Hello")
````

----
No `ZeroDivisionError` this time but the code still crashes due to a `TypeError` because the function expects a number and
we gave it a string.

So what do you do? Do you write an `if/else` construct for every known exception? What if the program fails
in a way that you did not anticipate?

Thankfully, there is a better way of handling exceptions in Python using the `try/except` construct. The syntax is:


```python
    try: 
        do_something()
    except Exception:
        handle the exception
    finally:
        always execute this part (e.g., clean-up)
```

For example we could write.

````python
def my_function3(a):
    try:
        return 2/a + float(a)**300                  # Will raise an exception if an error occurs
    except ZeroDivisionError:
        print("Argument must be non-zero!")         # This will only be returned if we divide by Zero
    except TypeError:
        print("Argument must be a number!")         # This will only be returned if we pass the wrong argument to the function.
    except Exception as e:
        print("Another error has occured", e)       # This will appear for any other Exception.
    finally:
        print("Final block. Always executed.")    # The content of the finally block is always executed at the end.
        print("I'm done! a = ", a)                  
        print()
        
my_function3(2)        
my_function3(0) 
my_function3("Hello") 
my_function3(1e300) 
````

<h1> 4. Anonymous functions </h1>
(aka $\lambda$ functions or lambda functions)

Sometimes you only want to use a function once or twice within a specific portion of your Python code and it may seem extravagant to define a function and give it a specific name. That is what __anonymous__ or *lambda* functions are for. They are also useful for defining functions that will be applied to every element of an array (we'll see this later).

Instead of 

```python 
    def square(x):
        return x**2
```
We can use:

````python
f = lambda x: x**2

print(f(2), f(9))
````

````python
f = lambda x, y: x*y if y > 0 else -2*x*y  

print(f(5, 6), f(5, -6), f(-5, 6))
````

<h4> <font color=MediumVioletRed> Note:</font> Any lambda function can be replaced by a "normal" function. </h4>

Anonymous functions are useful in some cases but if they become too complex it is probably a sign that a standard function should be used instead.

````python
# The previous lambda function could be coded as:

def f(x, y):
    if y > 0:
        return x*y
    else:
        return -2*x*y
    
print(f(5, 6), f(5, -6), f(-5, 6))
````

<h1> 5. Recursive functions </h1>
To finish on the topic (at least for now), let us mention that functions can be recursive (i.e., a function can call itself).

````python
def factorial(n):
    return n * factorial(n-1) if n > 0 else 1

print(factorial(5), factorial(-1))
````
